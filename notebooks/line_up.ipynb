```python
from ffl import espn, shark, models
import pandas as pd
```

```python
league_ids = [
    19470, # Road to Glory
    1317521, # Los Chivos
]

scorings = [
    10, # ESPN
    2, # PPR
]

segments = [
    636, # Week 9
    621, # Rest of 2018 season
]
```

```python
for i in league_ids:
    espn.update_projections(i)
```

```python
for sc in scorings:
    for se in segments:
        shark.update_projections(se, sc)
```

```python
league_id = 1317521
scoring = 2
segment = 636
position = "WR"
players = models.SharkProjections.query.filter_by(position=position, segment=segment, scoring=scoring).all()
df = pd.DataFrame(columns=["name", "status", "pts"])
for p in players:
    status = next((e.status for e in p.player.espn_projections if e.league_id == league_id), None)
    df = df.append({"name": p.player_name, "status": status, "pts": p.points}, ignore_index=True)
pd.options.display.max_rows = None
display(df.sort_values("pts", ascending=False))
```
